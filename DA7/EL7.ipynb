```python
import kagglehub
import os
from PIL import Image
import matplotlib.pyplot as plt

path = kagglehub.dataset_download("dataclusterlabs/indian-number-plates-dataset")
print("Path to dataset files:", path)


image_extensions = ('.jpg', '.jpeg', '.png')
all_images = []
for root, dirs, files in os.walk(path):
    for file in files:
        if file.lower().endswith(image_extensions):
            all_images.append(os.path.join(root, file))

if all_images:
    plt.figure(figsize=(12, 8))

    for i in range(min(3, len(all_images))):
        img = Image.open(all_images[i])
        plt.subplot(1, 3, i + 1)
        plt.imshow(img)
        plt.title(f"Sample {i+1}")
        plt.axis('off')
    plt.show()
else:
    print("No images found in the path. Please verify the download.")
```

```python
import re
import os
import cv2
import kagglehub
from PIL import Image
import sys

# Install pytesseract if not already available
try:
    import pytesseract
except ImportError:
    print("pytesseract not found, installing...")
    !{sys.executable} -m pip install pytesseract
    import pytesseract
    print("pytesseract installed and imported successfully.")

# --- Ensure all_images is available ---
if 'all_images' not in globals() or not all_images:
    print("Warning: 'all_images' not found or empty. Re-initializing from dataset.")
    try:
        path = kagglehub.dataset_download("dataclusterlabs/indian-number-plates-dataset")
        image_extensions = ('.jpg', '.jpeg', '.png')
        all_images = []
        for root, dirs, files in os.walk(path):
            for file in files:
                if file.lower().endswith(image_extensions):
                    all_images.append(os.path.join(root, file))
        if not all_images:
            print("Error: Failed to re-initialize 'all_images'. No images found.")
    except Exception as e:
        print(f"Error re-initializing 'all_images': {e}")
        all_images = [] # Ensure it's an empty list if re-initialization fails

# --- Ensure img_path is available ---
if 'img_path' not in locals() or img_path == '':
    if all_images:
        img_path = all_images[0]
        print(f"Warning: 'img_path' not defined or empty. Set to: {img_path}")
    else:
        img_path = ""
        print("Error: No images available to set 'img_path'.")

base_filename = os.path.basename(img_path)
ground_truth_data = {base_filename: 'MH12AB1234'}
ground_truth_text = ground_truth_data.get(base_filename, "")

# --- Ensure thresh_img is available ---
if 'thresh_img' not in locals():
    print("Warning: 'thresh_img' not defined. Attempting to re-generate.")
    if img_path and os.path.exists(img_path):
        original_img_cv2 = cv2.imread(img_path)
        if original_img_cv2 is None:
            print(f"Error: Could not load image at {img_path} to re-generate 'thresh_img'.")
            thresh_img = None # Indicate failure
        else:
            gray_img = cv2.cvtColor(original_img_cv2, cv2.COLOR_BGR2GRAY)
            # Use parameters from NH5ETNW0PNA2
            thresh_img = cv2.adaptiveThreshold(gray_img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 21, 10)
            print("'thresh_img' re-generated successfully.")
    else:
        print("Error: Cannot re-generate 'thresh_img' as 'img_path' is invalid or file does not exist.")
        thresh_img = None # Indicate failure

# --- Ensure extracted_text is available ---
if 'extracted_text' not in locals():
    print("Warning: 'extracted_text' not defined. Attempting to re-generate.")
    if thresh_img is not None:
        # Using PSM 3 as it was the last used in cell 0VUn004JQSjJ that had output
        extracted_text = pytesseract.image_to_string(thresh_img, config='--psm 3')
        print("'extracted_text' re-generated successfully using PSM 3.")
    else:
        print("Error: Cannot re-generate 'extracted_text' as 'thresh_img' is not available.")
        extracted_text = "" # Set to empty string to prevent further NameErrors

# Now perform the cleaning and comparison
cleaned_extracted_text = re.sub(r'[^a-zA-Z0-9]', '', extracted_text).strip().upper()
cleaned_ground_truth_text = re.sub(r'[^a-zA-Z0-9]', '', ground_truth_text).strip().upper()

print(f"Cleaned Ground Truth Text: {cleaned_ground_truth_text}")
print(f"Cleaned Extracted Text:    {cleaned_extracted_text}")
```

```python
from difflib import SequenceMatcher


is_match = (cleaned_extracted_text == cleaned_ground_truth_text)


similarity_ratio = SequenceMatcher(None, cleaned_ground_truth_text, cleaned_extracted_text).ratio()


print(f"\nComparison Result:")
print(f"Ground Truth: {cleaned_ground_truth_text}")
print(f"Extracted:    {cleaned_extracted_text}")
print(f"Exact Match:  {is_match}")
print(f"Similarity Ratio: {similarity_ratio:.2f}")
```

```python
character_accuracy = similarity_ratio
plate_accuracy = is_match

print(f"Character Accuracy (Similarity Ratio): {character_accuracy:.2f}")
print(f"Plate Accuracy (Exact Match): {plate_accuracy}")
```
